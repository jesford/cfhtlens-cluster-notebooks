# Playing with Clusters using Pandas
Here I'll collect notes and code related to storing and manipulating the CFHTLenS galaxy clusters using pandas dataframes.

```python
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
w1 = np.loadtxt('Clusters_W1.cat')
w1.shape
```

```python
df1 = DataFrame(w1, columns=['x0','x1','RA','DEC','z','sig','n200','r200','m200'])
```

```python
df1
```

## Plot cluster positions in W1
Note the gaps between separate 1 deg$^2$ megacam pointings...

```python
plt.plot(df1.RA,df1.DEC)
plt.axis('equal')
plt.xlabel('RA')
plt.ylabel('DEC')
```

## Select redshift column as a Series & plot histogram

```python
z = df1.z
type(z)
```

```python
zbins = np.arange(0.15,1.05,0.1)
plt.hist(z,bins=zbins)
plt.xlabel('Cluster Redshift')
```
